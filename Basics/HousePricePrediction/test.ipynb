```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
df = pd.read_csv("House Data India.csv")
df
```

```python
X = df["number of bedrooms"]
Y = df["Price"]

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.25)

# print(x_train.size, x_test.size, y_train.size, y_test.size)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(pd.DataFrame(x_train), y_train)
```

```python
y_pred = model.predict(pd.DataFrame(x_test))

from sklearn.metrics import accuracy_score, confusion_matrix
y_test = np.array(y_test)
# score = accuracy_score(x, y_pred.astype(int))
# score = confusion_matrix(x, y_pred.astype(int))
y_test, y_pred

onScreenVals = 20

plt.scatter(np.arange(onScreenVals), y_test[:onScreenVals], marker=".")
plt.scatter(np.arange(onScreenVals), y_pred[:onScreenVals], color="orange", marker="x")

for i in range(onScreenVals):
    plt.plot([i, i], [y_test[i], y_pred[i]], color='gray', linestyle='--', linewidth=1)
```
