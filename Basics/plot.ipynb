```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-4, 4, 1000)
sin = np.sin(x)
cos = np.cos(x)
plt.plot(x, sin)
plt.plot(x, cos)
plt.xticks(range(int(x[0]), int(x[-1])))
pass
diff = sin-cos
intersection_x = []
for i in range(len(x) - 1):
    if (sin[i] - cos[i]) * (sin[i + 1] - cos[i + 1]) < 0:
        intersection_x.append(i)

for i in intersection_x:
    plt.axvline(x[i], color="hotpink", linestyle="--")

```
