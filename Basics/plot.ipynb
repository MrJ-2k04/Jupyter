```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-4, 4, 1000)
sin = np.sin(x)
cos = np.cos(x)
plt.plot(x, sin, label="Sin")
plt.plot(x, cos, label="Cos")
plt.xticks(range(int(x[0]), int(x[-1])))
pass
diff = sin-cos
intersection_x = []
for i in range(len(x) - 1):
    if (sin[i] - cos[i]) * (sin[i + 1] - cos[i + 1]) < 0:
        intersection_x.append(i)

for i in intersection_x:
    plt.axvline(x[i], color="#333", linestyle="--", lw=1.5)

plt.fill_between(x, sin, color="red", alpha=.2)
plt.fill_between(x, cos, color="blue", alpha=.2)

plt.xlabel("Value")
plt.ylabel("Sin/Cos")
plt.legend(loc="upper right",shadow=True, fancybox=True)

```
